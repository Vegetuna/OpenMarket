```python
import pandas as pd
```

```python
# 방금 복사한 경로를 아래 '붙여넣기한_경로' 부분에 넣으세요
file_path = '../../data/데이터전체/카드승인매출정보.csv'

# 데이터 불러오기 (한글 깨짐 방지용 cp949)
df = pd.read_csv(file_path)

# 확인
print(df.head())
```

```python
import numpy as np

# ---------------------------------------------------------
# 가정: df라는 변수에 300만 개 전체 데이터가 이미 로드되어 있음
# 컬럼명 예시: '발급회원번호' (실제 컬럼명으로 수정 필요)
# ---------------------------------------------------------

# 1. 고유한 '발급회원번호' 목록 추출
unique_user_ids = df['발급회원번호'].unique()
print(f"전체 고유 회원 수: {len(unique_user_ids)}명")

# 2. 5만 명 무작위 샘플링 (비복원 추출)
# replace=False: 중복 없이 뽑기
# seed를 고정하면(np.random.seed) 실행할 때마다 같은 사람이 뽑힘
np.random.seed(42)
sampled_ids = np.random.choice(unique_user_ids, 50000, replace=False)

print(f"추출된 회원 수: {len(sampled_ids)}명")

# 3. 추출된 ID에 해당하는 데이터만 필터링 (isin 사용)
# '발급회원번호'가 sampled_ids 안에 있는 행만 남김
subset_df = df[df['발급회원번호'].isin(sampled_ids)]

# 4. 결과 확인
print("-" * 30)
print(f"추출 전 데이터 크기: {df.shape}")
print(f"추출 후 데이터 크기: {subset_df.shape}") # 약 300,000건 예상
print("-" * 30)

# 5. 파일로 저장 (필요시)
# subset_df.to_csv('sampled_data_300k.csv', index=False, encoding='cp949')
```

```python
subset_df.to_csv('승인매출정보_300k.csv', index=False, encoding='cp949')
```
