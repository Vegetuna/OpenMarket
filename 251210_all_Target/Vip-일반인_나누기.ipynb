```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/통합데이터/Final_merged_all_data.csv')
```

```python
df['VIP등급코드']
```

```python
# -------------------------------------------------------
# VIP와 일반인 데이터 분리
# -------------------------------------------------------
# vip코드 컬럼 기준
# '_' : 일반인
# 그 외 : VIP

df_normal = df[df['VIP등급코드'] == '_'].copy()
df_vip = df[df['VIP등급코드'] != '_'].copy()

# 결과 확인
print(f"전체 데이터 수: {len(df)}")
print(f"일반인(Normal) 수: {len(df_normal)}")
print(f"VIP 수: {len(df_vip)}")

# 검증: 두 그룹 합이 전체와 같은지 확인
if len(df) == len(df_normal) + len(df_vip):
    print("✅ 데이터 분리 검증 완료 (누락/중복 없음)")
else:
    print("⚠️ 경고: 데이터 분리 중 오차 발생")

print("\n[일반인 데이터 샘플]")
print(df_normal[['발급회원번호', 'VIP등급코드']].head())
print("\n[VIP 데이터 샘플]")
print(df_vip[['발급회원번호', 'VIP등급코드']].head())
```

```python
df_normal.to_csv('df_normal.csv', index=False)
df_vip.to_csv('df_vip.csv', index=False)
```

```python
df_normal.to_csv('일반인데이터.csv', index=False)
df_vip.to_csv('VIP데이터.csv', index=False)
```
