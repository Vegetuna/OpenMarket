```python
# 로드된 모든 DataFrame을 저장할 리스트
all_dfs = []

list = ['07','08','09','10','11','12']
for i in list:
    csv_file_path = f'/content/drive/MyDrive/원본 데이터/05.카드_잔액정보/2018{i}_잔액정보.csv'

    try:
        df = pd.read_csv(csv_file_path)
        all_dfs.append(df) # 로드된 DataFrame을 리스트에 추가
        print(f"'{csv_file_path}' 파일이 성공적으로 로드되었습니다.")

    except FileNotFoundError:
        print(f"오류: '{csv_file_path}' 파일을 찾을 수 없습니다. 경로를 확인해주세요.")
    except Exception as e:
        print(f"파일 로드 중 오류 발생: {e}")

print(f"총 {len(all_dfs)}개의 DataFrame이 로드되었습니다.")
# 예시: 첫 번째 로드된 DataFrame의 head를 보려면 display(all_dfs[0].head())를 사용하세요

# all_dfs 리스트에 있는 모든 DataFrame을 하나로 합칩니다.
# ignore_index=True를 설정하여 합쳐진 DataFrame의 인덱스를 재설정합니다.
combined_df_05 = pd.concat(all_dfs, ignore_index=True)

print(f"모든 DataFrame이 성공적으로 결합되었습니다. 결합된 DataFrame의 행 수는 {len(combined_df_05)} 입니다.")
display(combined_df_05.head())
```
