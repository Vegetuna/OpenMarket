```python

# 2. 파일 목록 정의 (실제 파일 경로로 수정해주세요)
file_list = [
    r"C:\Users\student\Desktop\card_data\01.카드 회원정보\201807_회원정보.csv",
    r"C:\Users\student\Desktop\card_data\02.카드 신용정보\201807_신용정보.csv",
    r"C:\Users\student\Desktop\card_data\03.카드 승인매출정보\201807_승인매출정보.csv",
    r"C:\Users\student\Desktop\card_data\04.카드 청구정보\201807_청구정보.csv",
    r"C:\Users\student\Desktop\card_data\05.카드 잔액정보\201807_잔액정보.csv",
    r"C:\Users\student\Desktop\card_data\06.카드 채널정보\201807_채널정보.csv",
    r"C:\Users\student\Desktop\card_data\07.카드 마케팅정보\201807_마케팅정보.csv",
    r"C:\Users\student\Desktop\card_data\08.카드 성과정보\201807_성과정보.csv"
]
```

```python

# 결과를 저장할 딕셔너리 (컬럼명 : [파일리스트])
# 하나의 컬럼이 여러 파일에 중복될 수도 있으므로 리스트로 저장
column_locations = {col: [] for col in target_columns}
found_columns = []
missing_columns = []

# 3. 파일 순회 및 검색
for file_path in file_list:
    try:
        # 핵심: nrows=0 으로 설정하여 헤더만 순식간에 읽어옴 (데이터 로드 X)
        # 한글 파일인 경우 encoding='cp949' 또는 'euc-kr' 필요할 수 있음
        df = pd.read_csv(file_path, nrows=0) 
        # 만약 엑셀이라면: pd.read_excel(file_path, nrows=0)
        
        file_columns = df.columns
        print(f"Checking file: {file_path} ...")

        # 타겟 컬럼이 현재 파일 컬럼에 있는지 확인 (요청하신 is in 스타일 로직)
        for target in target_columns:
            if target in file_columns:
                column_locations[target].append(file_path)
                if target not in found_columns:
                    found_columns.append(target)
                    
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```

```python

# 4. 결과 출력
print("\n" + "="*50)
print("검색 결과")
print("="*50)

# 찾은 컬럼 출력
for col in target_columns:
    files = column_locations[col]
    if files:
        print(f"[O] '{col}' -> {files}")
    else:
        missing_columns.append(col)

# 못 찾은 컬럼 출력 (혹시 오타가 있거나 파일이 누락된 경우 확인용)
print("\n" + "="*50)
print(f"발견되지 않은 컬럼: {len(missing_columns)}개")
print("="*50)
for col in missing_columns:
    print(f"[X] '{col}'")
```

```python
# 파일별로 어떤 컬럼이 있는지 담을 딕셔너리 생성 (1번~8번)
# file_list의 순서에 따라 인덱스를 매칭합니다.
file_num_to_cols = {i: [] for i in range(1, 9)}

# 아까 구한 결과(column_locations)를 뒤집어서 정리
for col, files in column_locations.items():
    for file_path in files:
        # file_list에서 해당 파일 경로가 몇 번째인지 찾음 (0부터 시작하므로 +1)
        try:
            idx = file_list.index(file_path) + 1
            file_num_to_cols[idx].append(col)
        except ValueError:
            pass # file_list에 없는 파일이면 패스

# 요청하신 포맷대로 출력
print("=" * 50)
for i in range(1, 9):
    print(f"{i}번컬럼")
    # 보기 좋게 정렬해서 출력 (선택사항)
    current_cols = sorted(file_num_to_cols[i])
    print(current_cols)
    print("-" * 50) # 구분선
```
